```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/abhijitchak103/iNeuron-Assignments/main/Python%20Questions/meteorites.csv")
```

```python
df["Year_at_which_Earth_Meteorite_was_hit"] = pd.to_datetime(df["Year_at_which_Earth_Meteorite_was_hit"], errors="coerce")
```

```python
df.info()
```

### Get all the Earth meteorites that fell before the year 2000

```python
import datetime as dt

meteorites_before_2000 = list(df[df["Year_at_which_Earth_Meteorite_was_hit"].dt.year < 2000]["Name_of_Earth_Meteorite"])
meteorites_before_2000
```

### Get all the earth meteorites co-ordinates who fell before the year 1970

```python
meteorite_coordinates_before_1970 = list(df[df["Year_at_which_Earth_Meteorite_was_hit"].dt.year < 1970]\
                                        ["Point_Coordinates"])
meteorite_coordinates_before_1970
```

### Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
meteor_mass_more_than_10000kg = list(df[df["Mass_of_Earth_Meteorite"] > 10000]["Name_of_Earth_Meteorite"])
meteor_mass_more_than_10000kg
```
