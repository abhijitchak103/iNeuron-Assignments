```python
# Import important libraries and modules

import requests, json
import pandas as pd
```

```python
# Save url as a local variable for easier usage

src = "https://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
```

```python
response = requests.get(src)

if response.status_code == 200:
    data = response.text
```

```python
json_data = json.loads(data)
```

```python
# Import pprint for proper visualization of dictionaries

import pprint
```

```python
# Use pprint.pprint to visualize structured dictionary

pprint.pprint(json_data)
```
